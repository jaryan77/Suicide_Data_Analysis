```python
import pandas as pd
import numpy as np
```

```python
file_path = "data/master.csv"

df = pd.read_csv(file_path)
```

```python
df.head(30)
```

```python
df = df.rename(columns={'country': 'Country', 'year': 'Year', 'sex': 'Gender', 'age': "Age Group", 'suicides_no': 'Suicides', 'population': 'Population', 'suicides/100k pop': 'Suicide per 100k', 'generation':'Generation'})
df.drop(columns=['country-year'])
df.head()
```

## General Numbers

```python
# Total Suicides

total = df['Suicides'].sum()
total_suicides = pd.DataFrame({"Total Suicides": [total]})

total_suicides
```

## Gender Analysis

```python
total_gender = df['Gender'].value_counts()
gender_percentage = total_gender/total * 100
gender_sum = pd.DataFrame({"Total": total_gender,
                         "Percentage": gender_percentage
                          })
#format %
gender_sum["Percentage"] = gender_sum["Percentage"].map("{:.2f}%".format)
gender_sum

#Nope
```
